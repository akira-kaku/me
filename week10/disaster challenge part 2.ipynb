```python
def get_disaster_summary(iso_code):
    # ISO
    country_data = emdat[emdat['ISO'] == iso_code]
    
    # 
    country_name = country_data['Country'].iloc[0]
    
    # 
    total_disasters = country_data.shape[0]
    
    # 
    total_deaths = country_data['Total Deaths'].sum()
    
    # 
    print(':', country_name)
    print(':', total_disasters)
    print(':', total_deaths)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# ---  ---
# 
DISASTER_TRANSLATION = {
    'Earthquake': '',
    'Flood': '',
    'Storm': '',
    'Drought': '',
    'Volcanic activity': '',
    'Mass movement (dry)': '',
    'Epidemic': '/',
    # 
}

# ---  ---
# 
try:
    plt.rcParams['font.family'] = 'sans-serif'
    plt.rcParams['font.sans-serif'] = ['Hiragino Kaku Gothic ProN', 'Yu Gothic', 'Meiryo', 'TakaoGothic', 'IPAexGothic', 'DejaVu Sans']
    plt.rcParams['axes.unicode_minus'] = False
except:
    print(" ")


def visualize_deadly_events(df, target_country, target_disaster_type, highlight_threshold=1000):
    """
    
    
    """
    
    # 
    df.columns = df.columns.str.strip()
    df_filtered = df[['Start Year', 'Country', 'Disaster Type', 'Total Deaths']].copy()
    
    # 'Total Deaths'
    df_filtered['Total Deaths'] = pd.to_numeric(df_filtered['Total Deaths'], errors='coerce')
    
    # ---  ---
    
    # 1. 
    filtered_data = df_filtered[
        (df_filtered['Country'] == target_country) &
        (df_filtered['Disaster Type'] == target_disaster_type)
    ].dropna(subset=['Total Deaths']) # NaN
    
    if filtered_data.empty:
        print(f" '{target_country}'  '{target_disaster_type}' ")
        return
        
    # 
    yearly_deaths = filtered_data.groupby('Start Year')['Total Deaths'].sum().reset_index()
    
    # ---  ---
    
    # 
    highlight_data = yearly_deaths[yearly_deaths['Total Deaths'] >= highlight_threshold]

    # ---  ---
    
    # 
    japanese_disaster_name = DISASTER_TRANSLATION.get(target_disaster_type, target_disaster_type)

    plt.style.use('seaborn-v0_8-whitegrid')
    fig, ax = plt.subplots(figsize=(14, 7))

    # 1. 
    ax.plot(
        yearly_deaths['Start Year'], 
        yearly_deaths['Total Deaths'], 
        label=' ()',
        color='gray', 
        linestyle='-', 
        linewidth=1.5
    )
    
    # 2. 
    if not highlight_data.empty:
        ax.scatter(
            highlight_data['Start Year'], 
            highlight_data['Total Deaths'], 
            label=f' ({highlight_threshold:,.0f})',
            color='red', 
            s=80,
            zorder=3
        )
        
        # 
        for index, row in highlight_data.sort_values(by='Total Deaths', ascending=False).head(3).iterrows():
            ax.annotate(
                f"{row['Total Deaths']:,.0f}", 
                (row['Start Year'], row['Total Deaths']),
                textcoords="offset points", 
                xytext=(0,10), 
                ha='center', 
                fontsize=9,
                color='darkred'
            )

    #  ()
    title = f"{target_country}{japanese_disaster_name} "
    ax.set_title(title, fontsize=16, pad=20)
    ax.set_xlabel('', fontsize=12)
    ax.set_ylabel(' ()', fontsize=12)
    
    # Y
    ax.get_yaxis().set_major_formatter(
        plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x)))
    )

    # 
    ax.legend(loc='upper left', fontsize=10)
    
    # X
    ax.xaxis.set_major_locator(plt.MaxNLocator(integer=True))
    
    plt.tight_layout()
    plt.show()

# ---  ---

# 
df_emdat = pd.read_csv('emdat (4).csv')

print("---  1: 1000 ---")
visualize_deadly_events(
    df=df_emdat, 
    target_country='Japan', 
    target_disaster_type='Earthquake', 
    highlight_threshold=1000
)

print("\n---  2: 20000 ---")
visualize_deadly_events(
    df=df_emdat, 
    target_country='China', 
    target_disaster_type='Flood', 
    highlight_threshold=20000
)
```
